```python
def plot_Q(Q,env):
    import matplotlib.pyplot as plt
    
    desc = env.unwrapped.desc.astype(str)

    num_states, num_actions = Q.shape
    grid_size = int(np.sqrt(num_states))

    action_arrows = {0: '←', 1: '↓', 2: '→', 3: '↑'}
    best_actions = np.argmax(Q, axis=1).reshape((grid_size, grid_size))

    tile_colors = np.zeros((grid_size, grid_size, 3))

    for i in range(grid_size):
        for j in range(grid_size):
            c = desc[i,j]
            if c == 'S':      # start = blue
                tile_colors[i,j] = [0.5, 0.8, 1.0]
            elif c == 'F':    # frozen = white
                tile_colors[i,j] = [1.0, 1.0, 1.0]
            elif c == 'H':    # hole = black
                tile_colors[i,j] = [0.0, 0.0, 0.0]
            elif c == 'G':    # goal = green
                tile_colors[i,j] = [0.0, 1.0, 0.0]

    plt.imshow(tile_colors, interpolation='none')

    for i in range(grid_size):
        for j in range(grid_size):
            arrow = action_arrows[best_actions[i,j]]
            if desc[i,j] == 'H': 
                color = 'white'
            else:
                color = 'black'
            plt.text(j, i, arrow, ha='center', va='center', fontsize=20, color=color)

    plt.xticks(range(grid_size))
    plt.yticks(range(grid_size))
    plt.show()
```

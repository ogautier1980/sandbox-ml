```python
import numpy as np
import matplotlib.pyplot as plt

def plot_mnist(dataset):
    targets = np.array(dataset.targets)
    random_indices = [np.random.choice(np.where(targets == digit)[0]) for digit in range(10)]
    imgs = [dataset[i] for i in random_indices]
    fig, axes = plt.subplots(1, 10, figsize=(15, 2))
    for i, (img, label) in enumerate(imgs):
        axes[i].imshow(img.squeeze(), cmap='gray')
        axes[i].set_title(str(label))
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def plot_mnist_with_predictions(model,dataset):
    model.eval()

    targets = np.array(dataset.targets)
    random_indices = [np.random.choice(np.where(targets == digit)[0]) for digit in range(10)]

    imgs = [dataset[i] for i in random_indices]

    fig, axes = plt.subplots(1, 10, figsize=(15, 2))

    for i, (img, true_label) in enumerate(imgs):
        x = img.to(device)

        with torch.no_grad():
            log_probs = model(x)
            probs = torch.exp(log_probs)
            pred_label = probs.argmax(dim=1).item()
            pred_prob  = probs[0, pred_label].item()

        axes[i].imshow(img.squeeze(), cmap='gray')
        axes[i].axis('off')
        axes[i].set_title(
            f"T:{true_label}\nP:{pred_label}\n({pred_prob:.2f})",
            fontsize=9
        )

    plt.tight_layout()
    plt.show()
```

## Tableaux (arrays)

```python
tab = np.array([0,1,2,3,4,5])
print(tab)
```

```python
print(tab.shape)
```

```python
element = tab[1]
print(element)
```

```python
serie = tab[-4:]
print(serie)
```

```python
mat = tab.reshape(3,2)
print(mat)
```

```python
print(mat.shape)
```

```python
ligne = mat[0,:]
colonne = mat[:,1]
print("La première ligne de la matrice est ",ligne, "et la deuxième colonne de la matrice est ", colonne)
```

```python
import matplotlib.pyplot as plt
```

```python
x = range(10)
y = np.array([20,17,15,12,11,10.5,5,4,2.7,-3])
z = np.linspace(-2,15,10)
plt.scatter(x,y)
plt.plot(x,y, label='courbe n°1')
plt.plot(x,z, label='courbe n°2')

plt.legend()
plt.xlabel('Axe x')
plt.ylabel('Axe y')
plt.title('Un titre')
plt.savefig('figure.png')
plt.show()
```

## Deuxième partie

Importation des librairies et des modules nécessaires

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

Importation du dataset et sauvegarde de la partie data/target

```python
cancer = datasets.load_breast_cancer()
x = cancer.data
y = cancer.target
```

Division des données en un ensemble d'entrainement et un ensemble de test

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42) 
```

Génération d'un arbre de décision, entrainement sur l'ensemble d'entrainement et affichage de l'arbre

```python
clf = DecisionTreeClassifier(max_leaf_nodes = 30)
clf.fit(x_train, y_train)
plot_tree(clf)
plt.show()
```

```python
clf = DecisionTreeClassifier(max_leaf_nodes = 20)
clf.fit(x_train, y_train)
plot_tree(clf)
plt.show()
```

<b>Attention :</b> l'arbre possède vingt feuilles. Il aurait pu atteindre jusqu'à 30, mais avec 20, il a déjà atteint une précision de 100 % sur les données d'entraînement. Ainsi, pour chaque valeur de max_leaf_nodes supérieure ou égale à 20, le résultat est le même

```python
clf = DecisionTreeClassifier(max_leaf_nodes = 15)
clf.fit(x_train, y_train)
plot_tree(clf)
plt.show()
```

Quel est l’impact du changement de la valeur de l’hyperparamètre max_leaf_nodes ? <b>l'arbre gagne en profondeur



Attention: 

<b>Ok, mais que se passe-t-il en termes de performances ?

Calcul des métriques pour évaluer la performance du modèle

```python
y_pred_test = clf.predict(x_test)
accuracy_test = accuracy_score(y_pred_test, y_test)
print("Accuracy de l'ensemble de test : ", accuracy_test)
precision = precision_score(y_pred_test, y_test)
print("Précision : ", precision)
recall = recall_score(y_pred_test, y_test)
print("Recall : ", recall)
f1 = f1_score(y_pred_test, y_test)
print("F1-score : ", f1)
conf_mat = confusion_matrix(y_pred_test, y_test, normalize='all')
print("Matrice de confusion :\n", conf_mat)
```

Calcul de l'accuracy sur l'ensemble d'entrainement

```python
y_pred_train = clf.predict(x_train)
accuracy_train = accuracy_score(y_pred_train, y_train)
print("Accuracy de l'ensemble d'entrainement : ", accuracy_train)
```

<b>Nous avons maintenant une accuracy d'entraînement de 100 % mais sur les données de test, elle est restée presque la même: 1.0 et 0.89

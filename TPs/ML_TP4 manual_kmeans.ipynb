```python
import matplotlib.pyplot as plt
from sklearn import datasets
from random import randint
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from IPython.display import clear_output
import time
import random
import numpy as np
```

```python
iris = datasets.load_iris()
data = iris.data[:,2:]
print(iris.feature_names)
plt.scatter(data[:,0],data[:,1])
```

```python
k = 2
centroids = []
for i in range(k):
    x = random.uniform(min(data[:,0]),max(data[:,0]))
    y = random.uniform(min(data[:,1]),max(data[:,1]))
    centroids.append((x,y))

plt.scatter(data[:,0],data[:,1])

plt.scatter([p[0] for p in centroids],[p[1] for p in centroids],marker="<")

plt.show()

for i in range(20):

    time.sleep(1)
    division = [[] for i in range(k)]
    for d in data:
        distances = []
        for i in range(k):
            centroid = centroids[i]
            distances.append((d[0]-centroid[0])**2 + (d[1]-centroid[1])**2)
        closest = np.argmin(distances)
        division[closest].append(d)

    new_centroids = []
    for i in range(k):
        x = np.mean([p[0] for p in division[i]])
        y = np.mean([p[1] for p in division[i]])
        new_centroids.append((x,y))


    plt.scatter(data[:,0],data[:,1])
        
    plt.scatter([p[0] for p in new_centroids],[p[1] for p in new_centroids],marker="<")
    clear_output(wait=True)
    plt.show()
    centroids = new_centroids
```

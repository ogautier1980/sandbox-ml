### Exercice 2 - Residual Block
```python
def forward(self, x):
    identity = x
    
    out = self.conv1(x)
    out = self.bn1(out)
    out = self.relu(out)
    
    out = self.conv2(out)
    out = self.bn2(out)
    
    if self.shortcut:
        identity = self.shortcut(x)
    
    out += identity  # Skip connection!
    out = self.relu(out)
    
    return out
```

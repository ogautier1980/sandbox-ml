# Chapitre 01 - DÃ©monstration : ProbabilitÃ©s et Statistiques

Ce notebook illustre les concepts probabilistes et statistiques essentiels au ML :
- Distributions de probabilitÃ©
- Statistiques descriptives
- ThÃ©orÃ¨me de Bayes
- Loi normale et TCL
- Covariance et corrÃ©lation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
np.random.seed(42)
```

## 1. Distributions de ProbabilitÃ© DiscrÃ¨tes

```python
# Distribution de Bernoulli
p = 0.7
bernoulli_rv = stats.bernoulli(p)

# GÃ©nÃ©rer des Ã©chantillons
samples_bernoulli = bernoulli_rv.rvs(size=1000)

print(f"Bernoulli(p={p}):")
print(f"  Moyenne empirique: {samples_bernoulli.mean():.3f} (thÃ©orique: {p})")
print(f"  Variance empirique: {samples_bernoulli.var():.3f} (thÃ©orique: {p*(1-p):.3f})")

# Distribution Binomiale
n, p = 10, 0.5
binomial_rv = stats.binom(n, p)

x_binom = np.arange(0, n+1)
pmf_binom = binomial_rv.pmf(x_binom)

# Visualiser
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Bernoulli
ax1.hist(samples_bernoulli, bins=[-0.5, 0.5, 1.5], density=True, alpha=0.7, edgecolor='black')
ax1.set_xticks([0, 1])
ax1.set_xlabel('Valeur', fontsize=12)
ax1.set_ylabel('FrÃ©quence', fontsize=12)
ax1.set_title(f'Bernoulli(p={p})', fontsize=14)

# Binomiale
ax2.bar(x_binom, pmf_binom, alpha=0.7, edgecolor='black')
ax2.set_xlabel('Nombre de succÃ¨s', fontsize=12)
ax2.set_ylabel('ProbabilitÃ©', fontsize=12)
ax2.set_title(f'Binomiale(n={n}, p={p})', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# Distribution de Poisson
lam = 3.5
poisson_rv = stats.poisson(lam)

x_poisson = np.arange(0, 15)
pmf_poisson = poisson_rv.pmf(x_poisson)

plt.figure(figsize=(10, 5))
plt.bar(x_poisson, pmf_poisson, alpha=0.7, edgecolor='black')
plt.axvline(lam, color='red', linestyle='--', linewidth=2, label=f'Î» = {lam}')
plt.xlabel('k', fontsize=12)
plt.ylabel('P(X = k)', fontsize=12)
plt.title(f'Poisson(Î»={lam})', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.show()

print(f"Poisson(Î»={lam}):")
print(f"  EspÃ©rance = Variance = {lam}")
```

## 2. Distributions Continues

```python
# Loi Uniforme
a, b = 0, 10
uniform_rv = stats.uniform(loc=a, scale=b-a)

# Loi Normale
mu, sigma = 5, 1.5
normal_rv = stats.norm(loc=mu, scale=sigma)

# Loi Exponentielle
lam_exp = 1.0
exp_rv = stats.expon(scale=1/lam_exp)

# Tracer les PDF
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Uniforme
x_unif = np.linspace(-1, 11, 1000)
axes[0].plot(x_unif, uniform_rv.pdf(x_unif), 'b-', linewidth=2)
axes[0].fill_between(x_unif, uniform_rv.pdf(x_unif), alpha=0.3)
axes[0].set_title(f'Uniforme({a}, {b})', fontsize=14)
axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].grid(True, alpha=0.3)

# Normale
x_norm = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
axes[1].plot(x_norm, normal_rv.pdf(x_norm), 'g-', linewidth=2)
axes[1].fill_between(x_norm, normal_rv.pdf(x_norm), alpha=0.3)
axes[1].axvline(mu, color='red', linestyle='--', label=f'Î¼ = {mu}')
axes[1].axvline(mu - sigma, color='orange', linestyle='--', alpha=0.7, label=f'Î¼ Â± Ïƒ')
axes[1].axvline(mu + sigma, color='orange', linestyle='--', alpha=0.7)
axes[1].set_title(f'Normale(Î¼={mu}, ÏƒÂ²={sigma**2})', fontsize=14)
axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('f(x)', fontsize=12)
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)

# Exponentielle
x_exp = np.linspace(0, 6, 1000)
axes[2].plot(x_exp, exp_rv.pdf(x_exp), 'r-', linewidth=2)
axes[2].fill_between(x_exp, exp_rv.pdf(x_exp), alpha=0.3)
axes[2].set_title(f'Exponentielle(Î»={lam_exp})', fontsize=14)
axes[2].set_xlabel('x', fontsize=12)
axes[2].set_ylabel('f(x)', fontsize=12)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 3. Loi Normale et RÃ¨gle 68-95-99.7

```python
# Loi normale standard N(0, 1)
mu, sigma = 0, 1
normal_rv = stats.norm(mu, sigma)

x = np.linspace(-4, 4, 1000)
pdf = normal_rv.pdf(x)

fig, ax = plt.subplots(figsize=(12, 6))

# Tracer la PDF
ax.plot(x, pdf, 'b-', linewidth=2, label='PDF')

# RÃ©gions
# 68% : Î¼ Â± Ïƒ
x_1sigma = x[(x >= -1) & (x <= 1)]
ax.fill_between(x_1sigma, normal_rv.pdf(x_1sigma), alpha=0.3, color='green', label='Î¼ Â± 1Ïƒ (68%)')

# 95% : Î¼ Â± 2Ïƒ
x_2sigma = x[(x >= -2) & (x <= -1)] 
ax.fill_between(x_2sigma, normal_rv.pdf(x_2sigma), alpha=0.3, color='yellow')
x_2sigma = x[(x >= 1) & (x <= 2)] 
ax.fill_between(x_2sigma, normal_rv.pdf(x_2sigma), alpha=0.3, color='yellow', label='Î¼ Â± 2Ïƒ (95%)')

# 99.7% : Î¼ Â± 3Ïƒ
x_3sigma = x[(x >= -3) & (x <= -2)] 
ax.fill_between(x_3sigma, normal_rv.pdf(x_3sigma), alpha=0.3, color='orange')
x_3sigma = x[(x >= 2) & (x <= 3)] 
ax.fill_between(x_3sigma, normal_rv.pdf(x_3sigma), alpha=0.3, color='orange', label='Î¼ Â± 3Ïƒ (99.7%)')

ax.axvline(0, color='black', linestyle='--', alpha=0.5)
for i in range(-3, 4):
    ax.axvline(i, color='gray', linestyle='--', alpha=0.3, linewidth=0.8)
    ax.text(i, -0.01, f'{i}Ïƒ', ha='center', fontsize=10)

ax.set_xlabel('x (Ã©carts-types)', fontsize=12)
ax.set_ylabel('DensitÃ©', fontsize=12)
ax.set_title('Loi Normale - RÃ¨gle 68-95-99.7', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.show()

# VÃ©rification numÃ©rique
print("ProbabilitÃ©s:")
print(f"  P(-1Ïƒ â‰¤ X â‰¤ 1Ïƒ) = {normal_rv.cdf(1) - normal_rv.cdf(-1):.4f} â‰ˆ 0.68")
print(f"  P(-2Ïƒ â‰¤ X â‰¤ 2Ïƒ) = {normal_rv.cdf(2) - normal_rv.cdf(-2):.4f} â‰ˆ 0.95")
print(f"  P(-3Ïƒ â‰¤ X â‰¤ 3Ïƒ) = {normal_rv.cdf(3) - normal_rv.cdf(-3):.4f} â‰ˆ 0.997")
```

## 4. ThÃ©orÃ¨me Central Limite (TCL)

```python
# Distribution non-normale (exponentielle)
true_dist = stats.expon(scale=2)

# ParamÃ¨tres
sample_sizes = [1, 5, 10, 30, 100]
n_simulations = 10000

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, n in enumerate(sample_sizes):
    # GÃ©nÃ©rer n_simulations moyennes de n Ã©chantillons
    sample_means = [true_dist.rvs(n).mean() for _ in range(n_simulations)]
    
    # Histogramme des moyennes
    axes[i].hist(sample_means, bins=50, density=True, alpha=0.7, edgecolor='black')
    
    # Superposer la loi normale thÃ©orique
    mu_theory = true_dist.mean()
    sigma_theory = true_dist.std() / np.sqrt(n)
    x_range = np.linspace(min(sample_means), max(sample_means), 100)
    axes[i].plot(x_range, stats.norm(mu_theory, sigma_theory).pdf(x_range),
                 'r-', linewidth=2, label=f'N({mu_theory:.2f}, {sigma_theory:.2f}Â²)')
    
    axes[i].set_title(f'Taille Ã©chantillon n = {n}', fontsize=13)
    axes[i].set_xlabel('Moyenne Ã©chantillonnale', fontsize=11)
    axes[i].set_ylabel('DensitÃ©', fontsize=11)
    axes[i].legend(fontsize=10)
    axes[i].grid(True, alpha=0.3)

# Distribution originale
x_orig = np.linspace(0, 10, 1000)
axes[5].plot(x_orig, true_dist.pdf(x_orig), 'b-', linewidth=2)
axes[5].fill_between(x_orig, true_dist.pdf(x_orig), alpha=0.3)
axes[5].set_title('Distribution Originale (Exponentielle)', fontsize=13)
axes[5].set_xlabel('x', fontsize=11)
axes[5].set_ylabel('DensitÃ©', fontsize=11)
axes[5].grid(True, alpha=0.3)

plt.suptitle('ThÃ©orÃ¨me Central Limite - Convergence vers la Normale', fontsize=16)
plt.tight_layout()
plt.show()
```

## 5. ThÃ©orÃ¨me de Bayes

```python
# Exemple : Test mÃ©dical
# Maladie rare : prÃ©valence = 0.1%
# Test prÃ©cis : sensibilitÃ© = 99%, spÃ©cificitÃ© = 99%

P_disease = 0.001  # P(Maladie)
P_positive_given_disease = 0.99  # P(Test+ | Maladie) - vrai positif
P_positive_given_healthy = 0.01  # P(Test+ | Sain) - faux positif

# P(Test+) = P(Test+|Maladie)*P(Maladie) + P(Test+|Sain)*P(Sain)
P_positive = (P_positive_given_disease * P_disease + 
              P_positive_given_healthy * (1 - P_disease))

# ThÃ©orÃ¨me de Bayes : P(Maladie | Test+)
P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive

print("=" * 60)
print("EXEMPLE : Test MÃ©dical (ThÃ©orÃ¨me de Bayes)")
print("=" * 60)
print(f"\nDonnÃ©es:")
print(f"  PrÃ©valence de la maladie: {P_disease*100:.2f}%")
print(f"  SensibilitÃ© du test (VP): {P_positive_given_disease*100:.1f}%")
print(f"  Taux de faux positifs: {P_positive_given_healthy*100:.1f}%")
print(f"\nRÃ©sultats:")
print(f"  P(Test positif) = {P_positive*100:.2f}%")
print(f"  P(Maladie | Test+) = {P_disease_given_positive*100:.2f}%")
print(f"\nâš ï¸  MÃªme avec un test Ã  99% de prÃ©cision,")
print(f"    si le test est positif, la probabilitÃ© d'Ãªtre malade")
print(f"    n'est que de {P_disease_given_positive*100:.2f}% !")
print(f"    C'est le paradoxe des tests sur maladies rares.")
print("=" * 60)
```

```python
# Visualiser l'effet de la prÃ©valence
prevalences = np.logspace(-4, -1, 100)  # De 0.01% Ã  10%
posteriors = []

for prev in prevalences:
    P_pos = P_positive_given_disease * prev + P_positive_given_healthy * (1 - prev)
    P_disease_pos = (P_positive_given_disease * prev) / P_pos
    posteriors.append(P_disease_pos)

plt.figure(figsize=(12, 6))
plt.semilogx(prevalences * 100, np.array(posteriors) * 100, 'b-', linewidth=2)
plt.axvline(P_disease * 100, color='red', linestyle='--', linewidth=2, 
            label=f'PrÃ©valence rÃ©elle ({P_disease*100:.2f}%)')
plt.axhline(P_disease_given_positive * 100, color='green', linestyle='--', linewidth=2,
            label=f'P(Maladie|Test+) = {P_disease_given_positive*100:.1f}%')
plt.xlabel('PrÃ©valence de la maladie (%)', fontsize=12)
plt.ylabel('P(Maladie | Test positif) (%)', fontsize=12)
plt.title('Impact de la PrÃ©valence sur le Diagnostic (Bayes)', fontsize=14)
plt.grid(True, alpha=0.3, which='both')
plt.legend(fontsize=11)
plt.show()
```

## 6. Statistiques Descriptives

```python
# GÃ©nÃ©rer des donnÃ©es
np.random.seed(42)
data = np.concatenate([
    np.random.normal(50, 10, 400),
    np.random.normal(80, 5, 100),  # DeuxiÃ¨me mode
    [5, 120]  # Outliers
])

# Statistiques
mean = np.mean(data)
median = np.median(data)
std = np.std(data, ddof=1)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1

print("Statistiques Descriptives:")
print(f"  Moyenne: {mean:.2f}")
print(f"  MÃ©diane: {median:.2f}")
print(f"  Ã‰cart-type: {std:.2f}")
print(f"  Q1 (25%): {q1:.2f}")
print(f"  Q3 (75%): {q3:.2f}")
print(f"  IQR: {iqr:.2f}")

# Visualisation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Histogramme
ax1.hist(data, bins=40, density=True, alpha=0.7, edgecolor='black')
ax1.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Moyenne = {mean:.1f}')
ax1.axvline(median, color='green', linestyle='--', linewidth=2, label=f'MÃ©diane = {median:.1f}')
ax1.axvline(mean - std, color='orange', linestyle='--', alpha=0.7, label=f'Î¼ Â± Ïƒ')
ax1.axvline(mean + std, color='orange', linestyle='--', alpha=0.7)
ax1.set_xlabel('Valeur', fontsize=12)
ax1.set_ylabel('DensitÃ©', fontsize=12)
ax1.set_title('Histogramme + Statistiques', fontsize=14)
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# Boxplot
ax2.boxplot(data, vert=True)
ax2.set_ylabel('Valeur', fontsize=12)
ax2.set_title('Boxplot (visualise mÃ©diane, Q1, Q3, outliers)', fontsize=14)
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

## 7. Covariance et CorrÃ©lation

```python
# GÃ©nÃ©rer des donnÃ©es corrÃ©lÃ©es
np.random.seed(42)
n = 200

# Trois paires avec corrÃ©lations diffÃ©rentes
correlations = [0.9, 0.0, -0.8]
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, rho in zip(axes, correlations):
    # GÃ©nÃ©rer des donnÃ©es avec corrÃ©lation rho
    mean = [0, 0]
    cov = [[1, rho], [rho, 1]]  # Matrice de covariance
    x, y = np.random.multivariate_normal(mean, cov, n).T
    
    # Calculer la corrÃ©lation empirique
    corr_empirical = np.corrcoef(x, y)[0, 1]
    
    # Scatter plot
    ax.scatter(x, y, alpha=0.6)
    
    # Ligne de rÃ©gression
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    x_line = np.linspace(x.min(), x.max(), 100)
    ax.plot(x_line, p(x_line), 'r-', linewidth=2, alpha=0.7)
    
    ax.set_xlabel('X', fontsize=12)
    ax.set_ylabel('Y', fontsize=12)
    ax.set_title(f'CorrÃ©lation Ï = {corr_empirical:.2f}', fontsize=14)
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')

plt.suptitle('Covariance et CorrÃ©lation', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Matrice de corrÃ©lation sur donnÃ©es rÃ©elles (exemple)
# GÃ©nÃ©rer un dataset synthÃ©tique
np.random.seed(42)
n_samples = 500

# Features corrÃ©lÃ©es
x1 = np.random.randn(n_samples)
x2 = 2 * x1 + np.random.randn(n_samples) * 0.5  # Fortement corrÃ©lÃ© avec x1
x3 = np.random.randn(n_samples)  # IndÃ©pendant
x4 = -1.5 * x1 + np.random.randn(n_samples) * 0.3  # NÃ©gativement corrÃ©lÃ© avec x1

# CrÃ©er DataFrame
df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Matrice de corrÃ©lation
corr_matrix = df.corr()

print("Matrice de CorrÃ©lation:")
print(corr_matrix)

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1)
plt.title('Matrice de CorrÃ©lation', fontsize=14)
plt.tight_layout()
plt.show()
```

## 8. Loi Normale MultivariÃ©e

```python
# GÃ©nÃ©rer des Ã©chantillons de loi normale multivariÃ©e
mean = [0, 0]
cov = [[1, 0.7], [0.7, 1.5]]  # Matrice de covariance

n_samples = 1000
samples = np.random.multivariate_normal(mean, cov, n_samples)

# Visualiser
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# Scatter plot
ax1.scatter(samples[:, 0], samples[:, 1], alpha=0.4, s=10)
ax1.set_xlabel('Xâ‚', fontsize=12)
ax1.set_ylabel('Xâ‚‚', fontsize=12)
ax1.set_title('Ã‰chantillons de Loi Normale BivariÃ©e', fontsize=14)
ax1.grid(True, alpha=0.3)
ax1.set_aspect('equal')

# Contour plot de la densitÃ©
from scipy.stats import multivariate_normal
rv = multivariate_normal(mean, cov)

x_range = np.linspace(-4, 4, 100)
y_range = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x_range, y_range)
pos = np.dstack((X, Y))
Z = rv.pdf(pos)

ax2.contourf(X, Y, Z, levels=20, cmap='viridis', alpha=0.7)
ax2.contour(X, Y, Z, levels=10, colors='black', alpha=0.3, linewidths=0.5)
ax2.set_xlabel('Xâ‚', fontsize=12)
ax2.set_ylabel('Xâ‚‚', fontsize=12)
ax2.set_title('Fonction de DensitÃ© de ProbabilitÃ©', fontsize=14)
ax2.grid(True, alpha=0.3)
ax2.set_aspect('equal')

plt.colorbar(ax2.contourf(X, Y, Z, levels=20, cmap='viridis', alpha=0.7), ax=ax2)
plt.tight_layout()
plt.show()

print("ParamÃ¨tres:")
print(f"  Moyenne: {mean}")
print(f"  Matrice de covariance:")
print(f"    {cov[0]}")
print(f"    {cov[1]}")
```
